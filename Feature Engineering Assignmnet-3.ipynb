```python
Q3. What is PCA (Principal Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.
PCA (Principal Component Analysis) is a technique used for dimensionality reduction by transforming the original features into a new set of uncorrelated features, called principal components. These components are ordered by the amount of variance they capture from the original dataset, and PCA seeks to reduce the number of dimensions while preserving as much information (variance) as possible.

Example: Consider a dataset with two features, X1 and X2, that are highly correlated. PCA will transform these into two principal components, PC1 and PC2, where PC1 captures most of the variance, and PC2 captures the rest. By retaining only PC1, you reduce the dimensionality from 2D to 1D while preserving most of the information.
```

```python
Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature Extraction? Provide an example to illustrate this concept.
PCA and feature extraction are related in that PCA creates new features (principal components) that are linear combinations of the original features. These new features capture the most important information (variance) from the original dataset. In feature extraction, the goal is to derive new, more informative features from the original ones, and PCA is one way to achieve this.

Example: Suppose you have a dataset with three features: X1, X2, and X3, where X1 and X2 are highly correlated. PCA would create principal components, and you might retain only PC1 and PC2, effectively extracting new features that reduce redundancy and noise.
```
